Find the total number of seconds from the given hours, minutes, and seconds.

```python
a = int(input("Enter the hours "))
b = int(input("Enter the minutes "))
c = int(input("Enter the second "))
d = a * 3600
e = b * 60
total_sec = d + e + c
print(f"Total_sec {a}hours {b}minutes {c}seconds gives us = {total_sec}")
```
