perfect square

```python
n = int(input("Enter the value of a "))
root = int(n**0.5)
if root*root ==n :
  print("perfect square")
else:
  print("not a perfect square")
```

Loops
while loop - works on a condition , execute till the condition is true

Print from 1 to 100

```python
n = 1
while (n<=100):
  print(n,end=",")
  n+=1
```

```python
n = 0
while (n<=15):
  print(f"Hello welcome to class {n}")
  n+=3
```

```python
n = 1
while (n<=10):
  print(f" 3 * {n} = {3*n}")
  n+=1
```

```python
n = int(input("Enter the value of n "))
i = 1
sum = 0
while(i<=n):
  sum = sum + i
  i+=1
print(f"The sum of the value n is {sum}")
```

```python
n = int(input("Enter the value of n "))
count = 0
while(n!=0):
  print(n)
  n = n//10
  count+=1
print(count)
```

```python
n = int(input("Enter the value of n "))
sum = 0
while(n!=0):
  d= n%10
  sum = sum + d
  n = n//10
print(sum)
```
