Login system:

If username correct

Check password

Else → invalid user

```python
correct_username = 'chiru'
correct_password = '123'
user_name = str(input("Enter username"))
password = str(input("Enter password"))
if (correct_username != user_name):
  print("Invalid user ")
elif (correct_password != password):
  print("Invalid password")
elif (correct_username == user_name and correct_password != password ):
  print("User name is correct and Password incorrect")
else :
  print("Login success")
```

Online shopping discount:

≥ 3000 → 10%
